```python
# 1st, sum up all the entities; 2nd, ; if time is enough, map the entities...
def count_ner(ner_rst, to_drop=[]):
    counter = Counter()
    for i in ner_rst:
        word, ne_type, *_ = i
        if ne_type in to_drop:
            continue
        counter[(word, ne_type)] += 1
    return counter
```
